```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("C:/Users/Home/Downloads/archive/diabetes.csv")
```

```python
y= df['Outcome']
X = df[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train_n= X_train.copy()
```

```python
X_test_n= X_test.copy()
```

```python
norm = MinMaxScaler().fit(X_train_n)
x_train_norm = norm.transform(X_train_n)
x_test_norm = norm.transform(X_test_n)
```

```python
x_train_norm_df = pd.DataFrame(x_train_norm)
x_test_norm_df = pd.DataFrame(x_test_norm)
```

```python
x_train_norm_df.columns = X_train_n.columns
```
